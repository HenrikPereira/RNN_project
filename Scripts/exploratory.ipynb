```python
# agrupar os nomes dos grupos de dados em listas
cat_cols = []
qnt_cols = []
date_cols = []
outc_cols = []
for col in train_df.columns:
    if col.startswith('Cat'):
        cat_cols.append(col)
    elif col.startswith('Quan'):
        qnt_cols.append(col)
    elif col.startswith('Date'):
        date_cols.append(col)
    elif col.startswith('Outcome'):
        outc_cols.append(col)
```

```python
date_and_outc = date_cols + outc_cols
```

```python
outc_c_cols = ['Outcome_M01',
 'Outcome_M02',
 'Outcome_M03',
 'Outcome_M04',
 'Outcome_M05',
 'Outcome_M06',
 'Outcome_M07',
 'Outcome_M08',
 'Outcome_M09',
 'Outcome_M10',
 'Outcome_M11',
 'Outcome_M12'
]
```

```python
train_dates_outc_df = train_df[date_and_outc].rename(columns=dict(zip(outc_cols, outc_c_cols))).reset_index().rename(columns={'index':'product', 'Date_2':'0', 'Date_1':'D_OfRelease'})
train_dates_outc_df.head()
```

```python
month = 30
it = 1
for i in outc_c_cols:
    train_dates_outc_df[str(int(i[-2:]))] = train_dates_outc_df['0'] + it * month
    it += 1
train_dates_outc_df['Outcome_M00'] = 0
train_dates_outc_df.head()
```

```python
ts_train_df = pd.melt(train_dates_outc_df, ['product', 'D_OfRelease'], [str(i) for i in range(0, 13)], 'Month', 'TimeValue')
ts_train_df['Month'] = ts_train_df['Month'].astype('int8')
ts_train_df['Outcome'] = pd.melt(train_dates_outc_df, ['product', 'D_OfRelease'], ['Outcome_M00'] + outc_c_cols, 'MonthOutcome', 'Outcome')['Outcome']
ts_train_df = ts_train_df.sort_values(by=['product', 'Month']).reset_index().drop('index', axis=1)
```

```python
ts_train_df
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(15,10))

ax = plt.axes(projection='3d')

x = ts_train_df['TimeValue']
y = ts_train_df['product']
z = ts_train_df['Outcome'].fillna(0)

ax.scatter3D(x, y, np.log(z), c=z, marker='o', cmap='Spectral_r', linewidth=0.5)
ax.set_xlabel('TimeValue')
ax.set_ylabel('Product')
ax.set_zlabel('Outcome')
ax.view_init(7, -100)
plt.show()

fig = plt.figure(figsize=(15,10))

ax = plt.axes(projection='3d')

ax.plot3D(x, y, np.log(z), c='blue', linewidth=0.5, alpha=0.8)
ax.set_xlabel('TimeValue')
ax.set_ylabel('Product')
ax.set_zlabel('Outcome')
ax.view_init(7, -100)
plt.show()
```

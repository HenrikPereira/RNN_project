```python
# agrupar os nomes dos grupos de dados em listas
cat_cols = []
qnt_cols = []
date_cols = []
outc_cols = []
for col in train_df.columns:
    if col.startswith('Cat'):
        cat_cols.append(col)
    elif col.startswith('Quan'):
        qnt_cols.append(col)
    elif col.startswith('Date'):
        date_cols.append(col)
    elif col.startswith('Outcome'):
        outc_cols.append(col)
```

```python
outc_c_cols = ['Outcome_M01',
 'Outcome_M02',
 'Outcome_M03',
 'Outcome_M04',
 'Outcome_M05',
 'Outcome_M06',
 'Outcome_M07',
 'Outcome_M08',
 'Outcome_M09',
 'Outcome_M10',
 'Outcome_M11',
 'Outcome_M12'
]
```

```python
train_outcomes_melt_df = pd.melt(train_df[outc_cols].rename(columns=dict(zip(outc_cols, outc_c_cols))).fillna(0).reset_index(), 'index', outc_c_cols, 'outcomes')
train_outcomes_melt_df.head()
```

```python
sb.lineplot(data=train_outcomes_melt_df, y='value', x='outcomes', hue='index')
plt.yscale('log')
plt.xticks(rotation=90)
plt.title('fill nan with zeroes')
plt.show()
```

```python
train_outcomes_melt_df = pd.melt(train_df[outc_cols].rename(columns=dict(zip(outc_cols, outc_c_cols))).fillna(train_df[outc_cols].median()).reset_index(), 'index', outc_c_cols, 'outcomes')
train_outcomes_melt_df.head()
```

```python
sb.lineplot(data=train_outcomes_melt_df, y='value', x='outcomes', hue='index')
plt.yscale('log')
plt.xticks(rotation=90)
plt.title('fill nan with median')
plt.show()
```
